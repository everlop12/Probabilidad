```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df1 = pd.read_csv("grupo1.csv")

print("--- Primeras 5 filas del Grupo 1 ---")
print(df1.head())

print("\n--- Descripción Estadística del Grupo 1 ---")
print(df1.describe())

plt.figure(figsize=(7, 4))
sns.histplot(df1['Calificacion'], bins=10, kde=True, color='skyblue')
plt.title('Histograma de Calificaciones Finales (Grupo 1)')
plt.xlabel('Calificación Final')
plt.ylabel('Frecuencia')
plt.show()
```

```python
correlacion_pearson = df1['HorasEstudio'].corr(df1['Calificacion'])
print(f"Coeficiente de Correlación (r) del Grupo 1: {correlacion_pearson:.4f}")

X = df1[['HorasEstudio']]
y = df1['Calificacion']

modelo = LinearRegression()
modelo.fit(X, y)

pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
R_cuadrado = modelo.score(X, y)

print(f"\n--- Ecuación de Regresión (Grupo 1) ---")
print(f"Intercepto (β₀): {intercepto:.4f}")
print(f"Pendiente (β₁): {pendiente:.4f}")
print(f"Ecuación: Calificación = {intercepto:.4f} + {pendiente:.4f} * HorasEstudio")
print(f"Coeficiente de Determinación (R²): {R_cuadrado:.4f}")
```

```python
y_pred = modelo.predict(X)

plt.scatter(X, y, color='blue', label='Datos Reales')
plt.plot(X, y_pred, color='red', linewidth=2, label=f'Línea de Regresión (R²={R_cuadrado:.2f})')
plt.title(f'Grupo 1: Regresión Lineal (r={correlacion_pearson:.2f})')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación Final')
plt.legend()
plt.grid(True)
plt.show()
```
