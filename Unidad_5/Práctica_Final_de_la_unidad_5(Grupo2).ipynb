```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

df2 = pd.read_csv("grupo2.csv")


print("--- Primeras 5 filas del Grupo 2 ---")
print(df2.head())

print("\n--- Descripción Estadística del Grupo 2 ---")

print(df2.describe())

plt.figure(figsize=(7, 4))
sns.histplot(df2['Calificacion'], bins=10, kde=True, color='lightcoral')
plt.title('Histograma de Calificaciones Finales (Grupo 2)')
plt.xlabel('Calificación Final')
plt.ylabel('Frecuencia')
plt.show()
```

```python
correlacion_pearson = df2['HorasEstudio'].corr(df2['Calificacion'])
print(f"Coeficiente de Correlación (r) del Grupo 2: {correlacion_pearson:.4f}")


X = df2[['HorasEstudio']]
y = df2['Calificacion']


modelo = LinearRegression()
modelo.fit(X, y)


pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
R_cuadrado = modelo.score(X, y)

print(f"\n--- Ecuación de Regresión (Grupo 2) ---")
print(f"Intercepto (β₀): {intercepto:.4f}")
print(f"Pendiente (β₁): {pendiente:.4f}")
print(f"Ecuación: Calificación = {intercepto:.4f} + {pendiente:.4f} * HorasEstudio")
print(f"Coeficiente de Determinación (R²): {R_cuadrado:.4f}")
```

```python
y_pred = modelo.predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Datos Reales')
plt.plot(X, y_pred, color='red', linewidth=2, label=f'Línea de Regresión (R²={R_cuadrado:.2f})')
plt.title(f'Grupo 2: Regresión Lineal (r={correlacion_pearson:.2f})')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación Final')
plt.legend()
plt.grid(True)
plt.show()
```

¿Explicar cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?


La pendiente ($\beta_1$) del Grupo 2 es 4.2227. Esto indica que, por cada hora adicional de estudio, la Calificación Final se predice que cambiará en 4.2227 puntos, en promedio.

¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y dirección de la relación entre las variables?


El coeficiente de correlación de Pearson ($r$) es 0.8620

Dirección: Como $r$ es positivo/negativo, la relación lineal es directa/inversa

Fuerza: El valor está cercano a 1/0, lo que sugiere una relación lineal fuerte/débil/moderada.

¿El modelo de regresión lineal es estadísticamente significativo?

La significancia estadística se evalúa con el p-valor. No obstante, si el valor de $R^2$ es alto (cercano a 1), es una fuerte indicación de que el modelo es probablemente significativo; es decir, la relación observada entre las horas de estudio y la calificación es poco probable que se deba al azar.
