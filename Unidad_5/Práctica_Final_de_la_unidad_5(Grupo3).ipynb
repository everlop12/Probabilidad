```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression


df3 = pd.read_csv("grupo3.csv")

print("--- Primeras 5 filas del Grupo 3 ---")
print(df3.head())

print("\n--- Descripción Estadística del Grupo 3 ---")
print(df3.describe())

plt.figure(figsize=(7, 4))
sns.histplot(df3['Calificacion'], bins=10, kde=True, color='green')
plt.title('Histograma de Calificaciones Finales (Grupo 3)')
plt.xlabel('Calificación Final')
plt.ylabel('Frecuencia')
plt.show()
```

```python
correlacion_pearson = df3['HorasEstudio'].corr(df3['Calificacion'])
print(f"Coeficiente de Correlación (r) del Grupo 3: {correlacion_pearson:.4f}")


X = df3[['HorasEstudio']]
y = df3['Calificacion']


modelo = LinearRegression()
modelo.fit(X, y)

pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
R_cuadrado = modelo.score(X, y)

print(f"\n--- Ecuación de Regresión (Grupo 3) ---")
print(f"Intercepto (β₀): {intercepto:.4f}")
print(f"Pendiente (β₁): {pendiente:.4f}")
print(f"Ecuación: Calificación = {intercepto:.4f} + {pendiente:.4f} * HorasEstudio")
print(f"Coeficiente de Determinación (R²): {R_cuadrado:.4f}")
```

```python
y_pred = modelo.predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Datos Reales')
plt.plot(X, y_pred, color='red', linewidth=2, label=f'Línea de Regresión (R²={R_cuadrado:.2f})')
plt.title(f'Grupo 3: Regresión Lineal (r={correlacion_pearson:.2f})')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación Final')
plt.legend()
plt.grid(True)
plt.show()
```

1.-¿Explicar cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?


El coeficiente de regresión es positivo y muy alto ($+10.3668$). Esto significa que, en promedio, por cada aumento de una hora en el tiempo de estudio, la Calificación Final de los estudiantes del Grupo 3 se predice que aumentará en aproximadamente $10.37$ puntos.

2.-¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y dirección de la relación entre las variables?


Valor: $r = 0.9472$

Dirección: El signo positivo ($+0.9472$) indica una relación directa (a más horas de estudio, mayor calificación)

Fuerza: El valor de $r$ está muy cercano a 1. Esto indica una relación lineal muy fuerte o casi perfecta.

3.-¿El modelo de regresión lineal es estadísticamente significativo?

Sí, el modelo es altamente significativo.

El Coeficiente de Determinación ($R^2$) es muy alto ($0.8972$). Un valor tan cercano a 1 proporciona evidencia muy sólida de que el tiempo de estudio es un predictor significativo de la calificación final en este grupo.
