```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_autos = pd.read_csv('autos2 (1).csv')

df_autos['price'] = pd.to_numeric(df_autos['price'], errors='coerce')
df_autos.dropna(subset=['price'], inplace=True)


min_price = df_autos['price'].min()
max_price = df_autos['price'].max()
moda_price = df_autos['price'].mode()[0]
mean_price = df_autos['price'].mean()
median_price = df_autos['price'].median()
std_price = df_autos['price'].std()

print("Estadísticas de precios:")
print(f"Mínimo: {min_price:.2f}")
print(f"Máximo: {max_price:.2f}")
print(f"Moda: {moda_price:.2f}")
print(f"Media: {mean_price:.2f}")
print(f"Mediana: {median_price:.2f}")
print(f"Desviación Estándar: {std_price:.2f}")

print("\nResumen descriptivo:")
print(df_autos['price'].describe())
```

```python
num_bins = int(np.log2(len(df_autos)) + 1)

plt.figure(figsize=(10, 6))
df_autos['price'].plot(kind='hist', bins=num_bins, color='lightblue', edgecolor='black')
plt.title('Histograma de Precios de Automóviles')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='body-style', y='price', data=df_autos, palette='viridis')
plt.title('Distribución de Precios por Tipo de Carrocería')
plt.xlabel('Tipo de Carrocería')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drive-wheels', y='price', data=df_autos, palette='plasma')
plt.title('Distribución de Precios por Tipo de Tracción')
plt.xlabel('Tipo de Tracción')
plt.ylabel('Precio')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='engine-location', y='price', data=df_autos, palette='coolwarm')
plt.title('Distribución de Precios por Ubicación del Motor')
plt.xlabel('Ubicación del Motor')
plt.ylabel('Precio')
plt.show()
```
