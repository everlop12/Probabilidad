```python
import scipy.stats as st

n = 1000
p = 0.001

print(f"--- Análisis con Distribución Binomial (n={n}, p={p}) ---")
print("-" * 60)

x1 = 0
prob1 = st.binom.pmf(x1, n, p)

print(f"1. Probabilidad de que ninguno sea defectuoso (X = {x1}):")
print(f"   P(X = {x1}) = {prob1:.4f}")
print(f"   Resultado: {prob1*100:.2f}%")

print("-" * 60)

x2 = 3
prob2 = st.binom.pmf(x2, n, p)

print(f"2. Probabilidad de que tres sean defectuosos (X = {x2}):")
print(f"   P(X = {x2}) = {prob2:.4f}")
print(f"   Resultado: {prob2*100:.2f}%")

print("-" * 60)

x3 = 4
prob3 = st.binom.pmf(x3, n, p)

print(f"3. Probabilidad de que cuatro sean defectuosos (X = {x3}):")
print(f"   P(X = {x3}) = {prob3:.4f}")
print(f"   Resultado: {prob3*100:.2f}%")
```
