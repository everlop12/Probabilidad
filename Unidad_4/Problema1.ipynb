```python
import scipy.stats as st

# --- Datos del Problema ---
media = 70
desviacion_estandar = 4
poblacion_total = 500
limite_inferior = 68
limite_superior = 72

prob_superior = st.norm.cdf(limite_superior, loc=media, scale=desviacion_estandar)
prob_inferior = st.norm.cdf(limite_inferior, loc=media, scale=desviacion_estandar)

probabilidad_rango = prob_superior - prob_inferior

cantidad_estudiantes = probabilidad_rango * poblacion_total

print(f"--- Análisis de Pesos ---")
print(f"Media : {media} kg, Desviación Estándar : {desviacion_estandar} kg")
print("-" * 30)

print(f"a. La probabilidad de estudiantes que pesan más de {limite_inferior} y menos de {limite_superior} kg:")
print(f"   Probabilidad: {probabilidad_rango:.4f} (o {probabilidad_rango*100:.2f}%)")

print(f"\nb. Cuantos estudiantes pesan más de {limite_inferior} y menos de {limite_superior} kg:")
print(f"   Cantidad (Valor exacto): {cantidad_estudiantes:.2f}")
print(f"   Cantidad (Redondeado): {round(cantidad_estudiantes)} estudiantes")
```
