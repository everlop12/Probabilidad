```python
import scipy.stats as st
import numpy as np

media_lambda = 5

print(f"--- Análisis con Distribución de Poisson (λ = {media_lambda}) ---")
print("-" * 50)


x_a = 2
prob_a = st.poisson.pmf(x_a, media_lambda)

print(f"a). ¿Cuál es la probabilidad de que el número de personas ingresadas sea dos (X = {x_a})?")
print(f"    P(X = {x_a}) = {prob_a:.4f}")
print(f"    Respuesta: La probabilidad de que ingresen 2 personas es del {prob_a*100:.2f}%.")

print("-" * 50)

x_b = 10
prob_acumulada_hasta_10 = st.poisson.cdf(x_b, media_lambda)
prob_exceda_10 = 1 - prob_acumulada_hasta_10

print(f"b). ¿Es probable que x exceda de 10? (P(X > {x_b}))")
print(f"    P(X ≤ {x_b}) = {prob_acumulada_hasta_10:.4f}")
print(f"    P(X > {x_b}) = 1 - P(X ≤ {x_b}) = {prob_exceda_10:.4f}")

print("\n    Explicación:")
if prob_exceda_10 < 0.05:
    print(f"    Dado que la probabilidad P(X > 10) es muy baja ({prob_exceda_10*100:.2f}%), NO es probable que X exceda de 10.")
else:
    print(f"    Dado que la probabilidad P(X > 10) es alta ({prob_exceda_10*100:.2f}%), SÍ es probable que X exceda de 10.")
```
